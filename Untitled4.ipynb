```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
skills= pd.read_csv('/content/top_skill.csv')
```

```python
skills.head()
```

```python
skills_counts = skills['skills'].value_counts()

# Plot the skills counts
plt.figure(figsize=(10, 6))
skills_counts.plot(kind='bar')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.title('Skill Counts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
